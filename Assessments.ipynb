```python
def sqrt(x):
   

    r = x  # number taking in is equal to r 
    precision = 0.00001  # used for making the square root more precise
   
    

    while abs(x - r * r) > precision:
     
   
        r = (r + x / r) / 2  # newtons method to find a better approximation to get the closest decimal
  
    
    squareRoot = format('%.0f'%r) # format the square root 
    
    
    root = str(squareRoot) # turn float into a string
    root=list(root)[::1]   # add to a list 
    
    print(root[0], end='') # print the first digit
    print(".", end='')     # print a decimal point 
    
    for i in range(100):
        i+=1
        print(root[i], end='')  # print out all other numbers after the 1. 
        
    return r
    

sqrt(2* 10 **200)  # create the number with 200 digits to get a more specific answer

```

## STDDEVP  vs STDDEV.S

### Packages


```python
import numpy as np
import matplotlib.pyplot as plt
```

### Data

### Create a test population by running the following command

```python
Population=np.random.randint(low=1, high=88,size=100)
print(Population)


```

### Get a sample of the population

```python
SamplePopulation = []
SamplePopulation.append(Population[0])
SamplePopulation.append(Population[60])
SamplePopulation.append(Population[5])
SamplePopulation.append(Population[90])
SamplePopulation.append(Population[25])
print(SamplePopulation)
```

### STDDEV.P

```python
STDEVp = np.sqrt(np.sum((Population - np.mean(Population))**2)/len(Population))

print("standard deviation of entire population is : %.2f"%STDEVp)
```

### STDDEV.S

```python
STDEVs = np.sqrt(np.sum((SamplePopulation - np.mean(SamplePopulation))**2)/len(SamplePopulation)-1)

print("standard deviation of a sample is  : %.2f"%STDEVs)


```

First we use <br>

## Research

* The STDEV.P function is used when your data represents the entire population.
* The STDEV.S function is used when your data is a sample of the entire population.
* The STDEV.S is used when you don't have all the data.
* This then creates an estimation of the STDEV.

### Read in data

```python
iris = datasets.load_iris()
```

```python
X = iris.data[:,[0,1,2,3,]] #Get the 4 columns Sepal length,width and petal length
```

### K-Means

### Elbow Method 
Used to find the number of clusters that is most suited to the data

```python
Error=[]
for i in range(1,10):
    kmeans= KMeans(n_clusters = i).fit(X)
    kmeans.fit(X)
    Error.append(kmeans.inertia_)
plt.plot(range(1,10),Error)
plt.title('Elbow Method')
plt.xlabel('No of clusters')
plt.show()
```

The elbow shape is formed at point 3 which tells us the number of clusters to use is 3

```python
kMean = KMeans(n_clusters=3)
y_kmeans = kMean.fit_predict(X)   #Predict the closest cluster each sample in X belongs to.
print(y_kmeans)
kMean.cluster_centers_
```
